```python
import random
import numpy as np
import numpy.matlib
import pandas as pd

csv_data = pd.read_csv('trade_new.csv')
level = [22, 23, 25, 27]
csv_data['type5'] = csv_data.apply(lambda x: int(str(x["pluno"])[:5]), axis=1)
csv_data['type4'] = csv_data.apply(lambda x: int(str(x["pluno"])[:4]), axis=1)
csv_data['type3'] = csv_data.apply(lambda x: int(str(x["pluno"])[:3]), axis=1)
csv_data['type2'] = csv_data.apply(lambda x: int(str(x["pluno"])[:2]), axis=1)
csv_data = csv_data.loc[(csv_data["type2"]==22) | (csv_data["type2"]==23) | (csv_data["type2"]==25) | (csv_data["type2"]==27), :]
csv_data
```

```python
from datetime import datetime, timedelta

csv_data["purTime"] = csv_data["sldatime"].apply(lambda x:datetime.strptime(x, "%Y-%m-%d %H:%M:%S"))
csv_data["weekday"] = csv_data["purTime"].apply(lambda x:x.weekday())
purTimeMax = max(csv_data["purTime"])
purTimeMin = min(csv_data["purTime"])
print(purTimeMin, purTimeMax)
```

```python
csv_data["week"] = csv_data["purTime"].apply(lambda x:int((x-purTimeMin).days/7))
csv_data["month"] = csv_data["purTime"].apply(lambda x:x.month-2)
csv_data["day"] = csv_data["purTime"].apply(lambda x:int((x-purTimeMin).days))
csv_data = csv_data.sort_values(["purTime"])
csv_data["bndno"].fillna(csv_data["pluno"], inplace=True)
csv_data["bndno"] = csv_data["bndno"].apply(lambda x:int(x))
csv_data
```

```python
days = csv_data["day"].drop_duplicates().values
# weeks = csv_data["pluno"].drop_duplicates().values
```

```python
def getTimeInfo(time, item):
    col = csv_data[time].drop_duplicates().values
    row = csv_data[item].drop_duplicates().values
    g_dp = csv_data["qty"].groupby([csv_data[time], csv_data[item]]).sum()
    new_dp = pd.DataFrame(0.0, index = row, columns = col)
    for i in g_dp.index.levels[0]:
        for j in g_dp[i].index:
            new_dp[i][j] = g_dp[i][j]
    return new_dp

day_pluno = getTimeInfo("day", "pluno")
week_pluno = getTimeInfo("week", "pluno")
month_pluno = getTimeInfo("month", "pluno")

day_bndno = getTimeInfo("day", "bndno")
week_bndno = getTimeInfo("week", "bndno")
month_bndno = getTimeInfo("month", "bndno")

day_type5 = getTimeInfo("day", "type5")
week_type5 = getTimeInfo("week", "type5")
month_type5 = getTimeInfo("month", "type5")

day_type4 = getTimeInfo("day", "type4")
week_type4 = getTimeInfo("week", "type4")
month_type4 = getTimeInfo("month", "type4")

day_type3 = getTimeInfo("day", "type3")
week_type3 = getTimeInfo("week", "type3")
month_type3 = getTimeInfo("month", "type3")

day_type2 = getTimeInfo("day", "type2")
week_type2 = getTimeInfo("week", "type2")
month_type2 = getTimeInfo("month", "type2")
```

```python
day_pluno.to_csv("p1/day_pluno.csv")
week_pluno.to_csv("p1/week_pluno.csv")
month_pluno.to_csv("p1/month_pluno.csv")

day_bndno.to_csv("p1/day_bndno.csv")
week_bndno.to_csv("p1/week_bndno.csv")
month_bndno.to_csv("p1/month_bndno.csv")

day_type5.to_csv("p1/day_type5.csv")
week_type5.to_csv("p1/week_type5.csv")
month_type5.to_csv("p1/month_type5.csv")

day_type4.to_csv("p1/day_type4.csv")
week_type4.to_csv("p1/week_type4.csv")
month_type4.to_csv("p1/month_type4.csv")

day_type3.to_csv("p1/day_type3.csv")
week_type3.to_csv("p1/week_type3.csv")
month_type3.to_csv("p1/month_type3.csv")

day_type2.to_csv("p1/day_type2.csv")
week_type2.to_csv("p1/week_type2.csv")
month_type2.to_csv("p1/month_type2.csv")
```

```python
def getTime2Info(time):
    col2 = csv_data[time].drop_duplicates().values
    g_dp2 = csv_data["qty"].groupby([csv_data[time], csv_data["pluno"], csv_data["bndno"], csv_data["type5"], csv_data["type4"], csv_data["type3"], csv_data["type2"]]).sum()
    new_dp2 = pd.DataFrame(g_dp2)
    new_dp2.reset_index(inplace=True)
    new_dp3 = csv_data[["pluno", "bndno", "type5", "type4", "type3", "type2"]].drop_duplicates().sort_values(["pluno", "bndno"])
    for i in col2:
        new_dp3[i] = 0.0
    for i, row in new_dp2.iterrows():
        new_dp3.loc[(new_dp3["pluno"]==row["pluno"])&(new_dp3["bndno"]==row["bndno"]), int(row[time])] = row["qty"]
    return new_dp3

day_pluno_bndno_type = getTime2Info("day")
week_pluno_bndno_type = getTime2Info("week")
month_pluno_bndno_type = getTime2Info("month")

day_pluno_bndno_type.to_csv("p1/day_pluno_bndno_type.csv")
week_pluno_bndno_type.to_csv("p1/week_pluno_bndno_type.csv")
month_pluno_bndno_type.to_csv("p1/month_pluno_bndno_type.csv")
day_pluno_bndno_type
```
